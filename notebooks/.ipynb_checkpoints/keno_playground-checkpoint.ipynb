```python
import numpy as np
import pandas as pd

keno_master_r = '../ai_lotto/data/keno/master/keno_master_r.pkl'
#keno_master_dr = '../ai_lotto/data/keno/master/keno_master_dr.pkl'
DF_R = pd.read_pickle(keno_master_r)
#DF_R = DF_R.set_index('draw_id');
#DF_DR = pd.read_pickle(keno_master_dr)
#DF_DR = DF_DR.set_index('draw_id');
DF_R
```

```python
row=1; row_end=2529015; col=1; col_end=20  
while (row <= row_end):    
    col=1;
    while(col <= col_end):
        DF_DR.at[row,'dr' + str(DF_R.at[row,'r'+str(col)])] = 1
        col+=1;
    row += 1;

DF_DR.to_pickle('../ai_lotto/data/keno/master/keno_master_drr.pkl')
```

```python
#create an empty draw df
def get_dep_columns():
        i=1; dps=[];
        while(i<=80):
            dps.append('dp'+str(i));
            i+=1;        
        return dps;
np_dp = np.zeros(shape=(2529015, 80), dtype=np.int8)
dp_cols = get_dep_columns()
DF_DP = pd.DataFrame(np_dp, columns=dp_cols, index=DF_R.index)
#df_dr = df_dr.set_index('draw_id')
DF_DP.to_pickle('../ai_lotto/data/keno/master/keno_master_dp.pkl')
DF_DP
```

```python
#start row1 with 1
DF_DP.loc[1] = DF_DP.loc[1] + 1;
DF_DP
```

```python
# now reset picked numbers for DRAW_1 to 0
row=1; row_end=2529015; col=1; col_end=20  
while (row <= row_end):    
    col=1;
    while(col <= col_end):
        #reset picked nums to 0
        DF_DP.at[row,'dp' + str(DF_R.at[row,'r'+str(col)])] = 0
        col+=1;
    #copy the current row to next row and increment it by 1
    DF_DP.loc[row+1] = DF_DP.loc[row] + 1;
    row += 1;

DF_DP.to_pickle('../ai_lotto/data/keno/master/keno_master_dpp.pkl')
```
